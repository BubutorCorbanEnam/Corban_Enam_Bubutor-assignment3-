### Create 2 separate dictionaries and join them as a new dataset

```python
dict1 = {'make': ['Alfa-Romero', 'mercedes-benz', 'audi'], 'number_of_doors': [2, 4, 6]}
dict2 = {'make': ['porsche', 'volvo', 'Nissan'], 'number_of_doors': [3, 1, 5]}
df1 = pd.DataFrame(dict1)
df2 = pd.DataFrame(dict2)
merged_df = pd.concat([df1, df2], ignore_index=True)

print(merged_df)
```
